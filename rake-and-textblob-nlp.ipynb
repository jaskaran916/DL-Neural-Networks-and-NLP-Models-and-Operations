```python
# -*- coding: utf-8 -*-
"""
Created on Wed Feb 12 11:46:12 2025

@author: student
"""
import nltk
nltk.download('stopwords')

text= "NLTK is a leading platform for building Python programs to work with human Languages and provides easy-to-use interfaces to over 50 corpora and lexical resources such as WordMark suite of text processing libraries for classification, tokenization, stemming, tagging, reasoning, wrappers for industrial-strength NLP libraries, and an active discussion forum."


##textblob

import textblob
from textblob import TextBlob
blob = TextBlob(text)

# Extract Noun Phrases (TextBlob does not have a full-fledged NER module)
print("TextBlob NER(Noun Phrases):")
print(blob.noun_phrases)


##RAKE package

from rake_nltk import Rake

rake=Rake()

rake.extract_keywords_from_text(text)
phrases=rake.get_ranked_phrases_with_scores()

print ("Extracted phrases:",phrases)

```
