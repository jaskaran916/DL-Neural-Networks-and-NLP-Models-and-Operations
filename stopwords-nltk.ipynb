```python
import nltk
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stopwd_nltk = stopwords.words('English')
```

```python
print(stopwd_nltk)
```

```python
print(len(stopwd_nltk))
```

```python
Text = 'Florence Nightingale was a nurse who saved many lives in the 19th century. She was named after the city of Florence in Italy, where her parents went after they got married in 1818'
```

```python
words=[word for word in Text.split()if word.lower() not in stopwd_nltk]
```

```python
words
```

```python
new_text = ' '.join(words)
```

```python
new_text
```

```python
print('Old Length: ', len(Text))
```

```python
print('New Length: ', len(new_text))
```
