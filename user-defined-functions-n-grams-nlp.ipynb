```python

def generate_ngrams(text, n):
    # Split the text into words
    words = text.split()
    
    # Generate n-grams
    ngrams = [tuple(words[i:i+n]) for i in range(len(words)-n+1)]
    
    return ngrams

# Input text
text = "I love natural language processing more than anyone else in the world."

# Get user input for n
n = int(input("Enter the value of n (for n-grams): "))

# Generate and print unigrams
unigrams = generate_ngrams(text, 1)
print("Unigrams:", unigrams)

# Generate and print bigrams
bigrams = generate_ngrams(text, 2)
print("Bigrams:", bigrams)

# Generate and print trigrams
trigrams = generate_ngrams(text, 3)
print("Trigrams:", trigrams)

# Generate and print n-grams for user-defined n
ngrams = generate_ngrams(text, n)
print(f"{n}-grams:", ngrams)
```
