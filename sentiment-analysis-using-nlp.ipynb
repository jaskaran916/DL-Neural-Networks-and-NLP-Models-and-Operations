```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Load the CSV file
file_path = r"/kaggle/input/sentimentdataset-csv/sentimentdataset.csv"
data = pd.read_csv(file_path)

# Check the first few rows and column names
print("Column names:", data.columns)
print(data.head())




```

```python
# Define function to analyze sentiment
def analyze_sentiment(text):
    analysis = TextBlob(str(text))  # Convert to string to avoid errors
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Assuming the text column is named 'text' (change if necessary)
text_column_name = 'Sentiment'  # Update this if the column has a different name

if text_column_name in data.columns:
    data['Sentiment'] = data[text_column_name].apply(analyze_sentiment)
else:
    print(f"Column '{text_column_name}' not found in the DataFrame.")
```

```python
# Save results to CSV
output_file_path = r"C:\Users\student\Desktop\sentiment_analysis_results.csv"
data.to_csv(output_file_path, index=False)
print("Sentiment analysis completed. Results saved to:", output_file_path)
```
